```python
#import keras.layers
#import keras.models

from keras.layers import Dense, Activation
from keras.models import Sequential
```

```python
import pandas
data_B651 = pandas.read_excel('/content/Regression.xlsx')
```

```python
y = data_B651 ['Earnings']
```

```python
x = data_B651 [['Cost', 'Grad', 'Debt', 'City']]
```

```python
B651_NNmodel = Sequential()

B651_NNmodel.add( Dense(32, activation= 'relu', input_dim = 4))
B651_NNmodel.add( Dense(32, activation= 'relu'))
B651_NNmodel.add( Dense(32, activation= 'relu'))
B651_NNmodel.add( Dense(1))

B651_NNmodel.compile( loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
x_train = x[0:93]
x_test = x[93:116]

y_train = y[0:93]
y_test = y[93:116]
```

```python
B651_NNmodel.fit(x_train,y_train,epochs=10000)
```
